```python
# Q = "code to process incoming documents and add them to a collection",
Q = "Definition of the WhereDocument class",
q_emb = col._embed(Q)
print(len(q_emb), len(q_emb[0]))
```

```python
# results = col.get(
#     limit=10,
#     where={'project': 'chroma'},
#     include=['documents', 'metadatas', 'embeddings']
# )

results = col.query(
    query_embeddings=q_emb,
    n_results=10,
    where={'project': 'chroma'},
    where_document={"$contains": "WhereDocument"}
)
```

```python
for key, value in results.items():
    if value != None:
        print(key, len(value[0]))
    else:
        print(key, value)
```

```python
for idx, m, dist, doc in zip(results['ids'][0], results['metadatas'][0], results['distances'][0], results['documents'][0]):
    print(idx)
    print(dist)
    print(json.dumps(m, indent=2))
    print('-' * 80)
    print(doc)
    print('=' * 80)
    print()
```
